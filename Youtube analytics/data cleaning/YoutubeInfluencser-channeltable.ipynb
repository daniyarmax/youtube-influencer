

*   Unique PK count

```python
unique_pk_count = df['channel_id'].value_counts().sum()
duplicate_pk_count = df['channel_id'].duplicated().sum()

print(unique_pk_count)
print(duplicate_pk_count)
```



*   Null rate per column



```python
null_rate_per_column = df.isnull().mean() * 100
null_rate_per_column
```

```python
numeric_columns = ['subscriber_count', 'total_views', 'video_count', 'external_links_count',
                   'proportion_videos_with_ad_signals', 'proportion_videos_with_external_links']

print("Name, max, min, median")
for col in numeric_columns:
  max_value = df[col].max()
  min_value = df[col].min()
  median_value = df[col].median()


  print(col, max_value, min_value, median_value)
```

```python
df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
```

```python
for col in numeric_columns:
  df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

```python
def parse_list(x):
  if pd.isna(x) or x == '':
    return []
  return [i.strip() for i in str(x).split('|')]
```

```python
string_columns = ['matched_ad_keywords', 'matched_store_domains', 'channel_external_links']

for col in string_columns:
  df[col] = df[col].apply(parse_list)
```

```python
df.head(5)
```

```python
mask = df['channel_title'] == 'Sponsor Block'
print(mask.sum())
```

```python
total_views_zero = df[df['total_views'] == 0].index
df = df.drop(total_views_zero)
```

```python
del df['data_source']
del df['parser_data_source']
```

```python
subs_less_than_5000 = df[df['subscriber_count'] < 5000].index
df = df.drop(subs_less_than_5000)
```

```python
df['country'] = df['country'].fillna('unknown')
```

```python
del df['niche']
```

```python
df['views_per_subscriber'] = df['total_views']/df['subscriber_count']
```

```python
social_cols = ['has_instagram', 'has_telegram', 'has_tiktok',
    'has_vk', 'has_twitter_x', 'has_facebook', 'has_discord']

df['ecosystem_score'] = df[social_cols].sum(axis=1)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('channels_cleaned.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('videos.csv')
df_channels = pd.read_csv('channels_cleaned_postgres.csv', sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[df['channel_id'].isin(df_channels['channel_id'])]
df
```

```python
df['video_id'].duplicated().sum()
```

```python
df = df.fillna('')
```

```python
df['published_at'] = df['published_at'].replace('', pd.NA)
```

```python
df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
```

```python
numeric_cols = ['duration_seconds', 'view_count', 'like_count', 'comment_count', 'external_links_count']
```

```python
for col in numeric_cols:
  df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
def parse_list(x):
  if pd.isna(x) or x == '':
    return []
  return [i.strip() for i in str(x).split('|')]
```

```python
cols_to_parse = ['tags', 'matched_ad_keywords', 'external_links']
```

```python
for col in cols_to_parse:
  df[col] = df[col].apply(parse_list)
```

```python
del df['data_source']
```

Min, Max and Median of numeric columns

```python
summary_table = df[numeric_cols].describe().T[['min', 'max', '50%']]
summary_table = summary_table.rename(columns={'50%': 'median'})
print(summary_table)
```

Mininum value of each column is 0


```python
zero_view = df['view_count'] <= 0
zero_like = df['like_count'] <= 0
zero_comment = df['comment_count'] <= 0
zero_duration_second = df['duration_seconds'] <= 0

print("Zero view video count:", zero_view.sum())
print("Zero like video count:", zero_like.sum())
print("Zero comment video count:", zero_comment.sum())
print("Zero duration seconds video count:", zero_duration_second.sum())
```
